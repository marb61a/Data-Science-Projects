#### Merging and Concatentaing Datasets

```python
# Vertically concatenates summer and winter
olympics = pd.concat([summer, winter], axis=0, keys=["Summer", "Winter"], names=["Edition"]).reset_index().drop(columns="level_1")
```

```python
olympics
```

```python
# Merge the new olympics dataset with parts of the dic dataset
olympics.merge(dic.iloc[:, :2], how="left", left_on="Country", right_on="Code").drop(columns=["Code"])
```
