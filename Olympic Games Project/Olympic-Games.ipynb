### Hypothesis Testing
#### Does GDP, Population infulence Olympic success

```python
medals_per_country
```

```python
medals_per_country.drop(columns=["Gold", "Silver", "Bronze"], inplace=True)
```

```python
# Creates the correlation matrix, pearson is the most used
medals_per_country.corr(method="pearson")
```

```python
medals_per_country.corr(method="spearman")
```

```python
medals_per_country.rank(ascending=False).corr(method="pearson")
```

#### Hypothesis 1 - There is no relationship between Total Medals and Population
#### A correlation value = 0 means no correlation, up to 1 means that the is a correlation, down to -1 indicates a negative correlation

```python
import scipy.stats as stats
```

```python
# There are some differences between Pearson and Spearman which makes Spearman the better choice in 
# these circumstances
stats.spearmanr(medals_per_country.Total, medals_per_country.Population, nan_policy="omit")
```

```python
r, p_value = stats.spearmanr(medals_per_country.Total, medals_per_country.Population, nan_policy="omit")
```

```python
r
```

```python
round(p_value)
```

#### This hypothesis is rejected as there is a significant positive relationship between Population and Total Medals

#### Hypothesis 2 - There is a significant and positive relationship between Total Medals and GDP per Capita

```python
r, p_value = stats.spearmanr(medals_per_country.Total, medals_per_country.GDP, nan_policy="omit")
```

```python
r
```

```python
p_value
```

#### Hypothesis 2 is rejected as there is a positive and significant relationship between GDP Per Capita and Total Medals

#### Hypothesis 3 - There is no relationship between Total Medals and Participations (How many Olympic Games has the country competed in)

```python
medals_per_country.head(50)
```

```python
r, p_value = stats.spearmanr(medals_per_country.head(50).Total, medals_per_country.head(50).Total_Games, nan_policy="omit")
```

```python
r
```

```python
p_value
```

#### Hypothesis 3 is rejected as there is a significant and positive relationship between Total Medals and Participations in games

### Aggregating and Ranking

```python
olympics
```

```python
# Compare the summer and winter editions
sum_vs_win = pd.crosstab(olympics.Country, olympics.Edition)
```

```python
sum_vs_win
```

```python
# Count and compare male and female athletes
men_vs_women = pd.crosstab(olympics.Country, olympics.Gender)
men_vs_women
```

```python
# Count and compare medals by categories
medals_by_cats = pd.concat([sum_vs_win, men_vs_women], axis=1)
medals_by_cats
```

```python
# Get category totals
medals_by_cats["Total"] = medals_by_cats.Summer + medals_by_cats.Winter
medals_by_cats
```

```python
# Sort category values
medals_by_cats.sort_values("Total", ascending=False, inplace=True)
medals_by_cats
```
